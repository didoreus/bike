```python
import dlt
from pyspark.sql.functions import *

catalog = spark.conf.get('catalog')
```

```python
@dlt.table(
    name = f'{catalog}.gold.daily_station_performance',
    comment = 'This table contains citibike data with trip duration in minutes')
def gold_daily_station_performance():
    df = dlt.read(f'{catalog}.silver.silver_jc_citibike')
    df = df.groupBy('trip_start_date').agg(max(col('trip_duration_mins')).alias('max_trip_duration_mins'), 
                                                          min(col('trip_duration_mins')).alias('min_trip_duration_mins'),
                                                          avg(col('trip_duration_mins')).cast('decimal(10,2)').alias('avg_trip_duration_mins'),
                                                          count('ride_id').alias('total_trips'))
    return df
```

```python
@dlt.table(
    name = f'{catalog}.gold.daily_ride_summary',
    comment = 'This table contains citibike data with trip duration in minutes'
)
def gold_daily_ride_summary():
    df = dlt.read(f'{catalog}.silver.silver_jc_citibike')
    df = df.groupBy('trip_start_date').agg(avg(col('trip_duration_mins')).cast('decimal(10,2)').alias('avg_trip_duration_mins'),count('ride_id').alias('total_trips'))
    return df
```
