```python
from pyspark.sql.functions import *

pipeline_id = dbutils.widgets.get('pipeline_id')
run_id = dbutils.widgets.get('run_id')
task_id = dbutils.widgets.get('task_id')
processed_timestamp = dbutils.widgets.get('processed_timestamp')
catalog = dbutils.widgets.get('catalog')

df_bronze = spark.read.format('csv').option('header', 'True').load(f'/Volumes/{catalog}/landing/operational_data/JC-202503-citibike-tripdata.csv')
df_bronze = df_bronze.select( col('_metadata.file_path').alias('file_path'),col('_metadata.file_modification_time').alias('load_time'),'*')
#display(df_bronze.limit(10))
```

```python
df_bronze = df_bronze.withColumn('metadata', 
                                 create_map(lit('pipeline_id'), lit(pipeline_id),
                                            lit('run_id'), lit(run_id),
                                            lit('task_id'), lit(task_id),
                                            lit('processed_timestamp'), lit(processed_timestamp)))
```

```python
df_bronze.write.mode('overwrite').option('overwriteSchema', 'True').saveAsTable(f'{catalog}.bronze.jc_citibike')
```
