```python
from pyspark.sql.functions import *

catalog = dbutils.widgets.get('catalog')


df = spark.read.table(f'{catalog}.silver.jc_citibike')
daily_ride_summary_df = df.groupBy('trip_start_date').agg(max(col('trip_duration_mins')).alias('max_trip_duration_mins'), 
                                                          min(col('trip_duration_mins')).alias('min_trip_duration_mins'),
                                                          avg(col('trip_duration_mins')).cast('decimal(10,2)').alias('avg_trip_duration_mins'),
                                                          count('ride_id').alias('total_trips'))
```

```python
daily_station_performance = df.groupBy('trip_start_date', 'start_station_name').agg(avg(col('trip_duration_mins')).cast('decimal(10,2)').alias('avg_trip_duration_mins'), 
                                                                                    count('ride_id').alias('total_trips'))

```

```python
daily_ride_summary_df.write.mode('overwrite').option('overwriteSChema', 'True').saveAsTable(f'{catalog}.gold.daily_ride_summary')
daily_station_performance.write.mode('overwrite').option('overwriteSChema', 'True').saveAsTable(f'{catalog}.gold.daily_station_performance')
```
